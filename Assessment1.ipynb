Step 19. Write a lambda function to convert a string column to a numerical column 


```python

d = {'M' : 0 , 'F' : 1}
#dict to map the gender values to int

func = lambda x : d[x]
#lambda fun to convert string to int values


users['gender'] = users['gender'].apply(func)
#converting the column gender from string to int 
```
